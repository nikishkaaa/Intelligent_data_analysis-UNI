```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
# Мережа Петрі: місця і переходи
class PetriNet:
    def __init__(self):
        self.places = {}
        self.transitions = {}
        self.arcs = []

    def add_place(self, place_name):
        self.places[place_name] = 0

    def add_transition(self, transition_name):
        self.transitions[transition_name] = []

    def add_arc(self, from_place, to_place):
        self.arcs.append((from_place, to_place))

    def fire_transition(self, transition_name):
        for from_place, to_place in self.arcs:
            if from_place == transition_name:
                if self.places[from_place] > 0:  # Якщо є токени
                    self.places[from_place] -= 1
                    self.places[to_place] += 1
```

```python
pn.places['Вибір спеціальності'] = 1
```

```python
# Моделюємо процес
pn.fire_transition('Перехід 1: Вибір спеціальності')
pn.fire_transition('Перехід 2: Пошук інформації')
pn.fire_transition('Перехід 3: Перевірка вимог')
pn.fire_transition('Перехід 4: Подача документів')
pn.fire_transition('Перехід 5: Складання іспитів')
pn.fire_transition('Перехід 6: Очікування результатів')

print(pn.places)
```

```python
G = nx.DiGraph()

for place in pn.places:
    G.add_node(place)

for transition in pn.transitions:
    G.add_node(transition)

for from_place, to_place in pn.arcs:
    G.add_edge(from_place, to_place)


plt.figure(figsize=(12, 12))
nx.draw(G, with_labels=True, node_size=3000, node_color='skyblue', font_size=10, font_weight='bold', arrows=True)
plt.title("Мережа Петрі: Процес вступу до університету")
plt.show()
```
