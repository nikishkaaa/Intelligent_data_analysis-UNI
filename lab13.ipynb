```python
import json
from lxml import html
import requests
from datetime import datetime
```

```python
headers = {
    'Accept': 'text/html,application/xhtml+xml,application/xml;q=0.9,image/avif,image/webp,image/apng,*/*;q=0.8,application/signed-exchange;v=b3;q=0.7',
    'Accept-Language': 'ru-RU,ru;q=0.9,en-US;q=0.8,en;q=0.7,uk;q=0.6',
    'Cache-Control': 'max-age=0',
    'Connection': 'keep-alive',
    'Referer': 'https://jobs.halliburton.com/search/?createNewAlert=false&q=&locationsearch=',
    'Sec-Fetch-Dest': 'document',
    'Sec-Fetch-Mode': 'navigate',
    'Sec-Fetch-Site': 'same-origin',
    'Sec-Fetch-User': '?1',
    'Upgrade-Insecure-Requests': '1',
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/121.0.0.0 Safari/537.36',
    'sec-ch-ua': '"Not A(Brand";v="99", "Google Chrome";v="121", "Chromium";v="121"',
    'sec-ch-ua-mobile': '?0',
    'sec-ch-ua-platform': '"Windows"',
}

params = {
    'q': '',
    'sortColumn': 'referencedate',
    'sortDirection': 'desc',
    'startrow': '25',
}

response = requests.get('https://jobs.halliburton.com/search/', params=params, headers=headers)
```

```python
JOBS_XPATH = '//span[contains(@class,"visible-phone")]/a[contains(@class, "jobTitle-link")]'
tree = html.fromstring(response.content)
jobs = tree.xpath(JOBS_XPATH)
for job in jobs:
  title = job.text
  url = f"https://jobs.halliburton.com{job.get('href')}"
  job_id = url.split("/")[-2]
  response = requests.get(url)
  DATE_XPATH = '//meta[contains(@itemprop, "datePosted")]/@content'
  LOCATION_XPATH = '//span[contains(@class, "jobGeoLocation")]/text()'
  EMPLOYMENT_TYPE= "//p/b[contains(text(), 'Part Time:')]/following-sibling::text()"
  DESCRIPTION_XPATH = '//span[contains(@itemprop, "description")]'
  tree = html.fromstring(response.content)
  date_posted = datetime.strptime(tree.xpath(DATE_XPATH)[0], "%a %b %d %H:%M:%S %Z %Y")
  location = tree.xpath(LOCATION_XPATH)[0].strip()
  if tree.xpath(EMPLOYMENT_TYPE):
    employment_type = tree.xpath(EMPLOYMENT_TYPE)[0].strip()
  else:
    employment_type = None
  description = tree.xpath(DESCRIPTION_XPATH)
  country = location.split(", ")[-2]
  if len(location.split(", ")[-1])>2:
    country = location.split(", ")[-2]
  else:
    country = location.split(", ")[-1]
  company = 'halliburton'
  print("Job ID:", job_id)
  print("Title:", title)
  print("Company:", company)
  print("Employment Type:", employment_type)
  print("Location:", location)
  print("Country:", country)
  print("Date Posted:", date_posted)
  print(description)
  print(url)
```
